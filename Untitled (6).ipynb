# Data Wrangling
Here we will perform the Data Cleaning process by weeding out irrelevant and incorrect infotmation.

### Completeness

```python
# Here we are checking for missing values 
aviation_data.isna().sum()
```

```python
# Fill missing numeric values with the median
numeric_cols = aviation_data.select_dtypes(include=['float64', 'int64']).columns
aviation_data[numeric_cols] = aviation_data[numeric_cols].fillna(aviation_data[numeric_cols].median())

# Fill missing categorical values with the mode
categorical_cols = aviation_data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    aviation_data[col] = aviation_data[col].fillna(aviation_data[col].mode()[0])

print("Remaining Missing Values:\n", aviation_data.isnull().sum())
```

The data has no missing values

```python
# checking for duplicates
aviation_data.duplicated().sum()
```

# Explatory Analysis


After cleaning the data and eliminating all utility we proceed to analyzing it. It will be conducted in three steps 

## Uniraviate Analysis

### Categorical variable

```python
# Strip leading/trailing spaces from column names
aviation_data.columns = aviation_data.columns.str.strip()

# Print the first few rows to understand the data structure
print(aviation_data.head())

# List of required columns
relevant_columns = ['Number.of.Engines', 'Aircraft.Category']

# Check if all required columns are present (case insensitive)
aviation_data_columns_lower = [col.lower() for col in aviation_data.columns]
relevant_columns_lower = [col.lower() for col in relevant_columns]

if not all(col in aviation_data_columns_lower for col in relevant_columns_lower):
    missing_columns = [col for col in relevant_columns if col.lower() not in aviation_data_columns_lower]
    print(f"One or more of the required columns are missing: {missing_columns}")
    print(f"Available columns are: {aviation_data.columns.tolist()}")
else:
    # Normalize the column names to match the required columns
    column_mapping = {col.lower(): col for col in aviation_data.columns}
    normalized_columns = [column_mapping[col.lower()] for col in relevant_columns_lower]
    
    # Plot 1: Histogram of Aircraft Category Frequency
    plt.figure(figsize=(10, 6))
    aviation_data[normalized_columns[1]].value_counts().plot(kind='bar', color='skyblue')
    plt.xlabel('Aircraft Category')
    plt.ylabel('Frequency')
    plt.title('Frequency of Aircraft Categories')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()

    # Plot 2: Histogram of Number of Engine Frequency
    plt.figure(figsize=(10, 6))
    aviation_data[normalized_columns[0]].value_counts().sort_index().plot(kind='bar', color='lightcoral')
    plt.xlabel('Number of Engines')
    plt.ylabel('Frequency')
    plt.title('Frequency of Number of Engines')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()
```

The histograms above show the frequency of aircrafts category and number of engines.

```python
# Distribution of
# Limit the number of unique values to plot for each categorical variable
max_unique_values = 5
catVariables = aviation_data.select_dtypes(include='object')
catColNames = list(catVariables.columns)

# Filter columns with fewer unique values than the limit
filtered_catColNames = [col for col in catColNames if catVariables[col].nunique() <= max_unique_values]

# Create subplots for filtered categorical variables
fig, axes = plt.subplots(ncols=3, nrows=(len(filtered_catColNames) // 3) + 1, figsize=(20, 15), sharey=True)
for idx, ax in enumerate(axes.ravel()):
    if idx < len(filtered_catColNames):
        variable = filtered_catColNames[idx]
        sns.countplot(data=catVariables, x=variable, ax=ax, color='salmon')
        ax.set(ylabel=' ', yticklabels=[])
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
        ax.set_title(variable)

fig.suptitle('Distribution of Categorical Variables (Limited to 20 Unique Values)')
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to make room for the title
plt.show()
```

The subplots above shows the distribution of different variables

### Numeric Variable

```python
# Select numerical variables
numColumns = aviation_data.select_dtypes(exclude='object')
numColNames = list(numColumns.columns)
# Create subplots for numerical variables
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
for col, ax in zip(numColNames, axes.ravel()):
    sns.boxplot(data=numColumns, y=col, ax=ax, color='green')
    ax.set_title(col)
fig.suptitle('Boxplots of Numerical Columns')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

The boxplot above show the distribution of the numeric data values in our dataset. There boxplot indicates the annormalies found our dataset for example total minor injuries has one annormaly.

```python
#The relationship between number of engines and total fatal injuries

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot
sns.scatterplot(
    x='total.fatal.injuries', 
    y='number.of.engines', 
    hue='aircraft.category', 
    data=aviation_data, 
    ax=ax, 
    marker='o'
)

# Display the plot
plt.show()
```

From the scatterplot above it is safe to say that aircrafts that have one engine have had more fatal injuries occuring from incidents and accidents.



```python
# Correlation Matrix
numVariables = aviation_data.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numVariables.corr()

# Plot the correlation matrix
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

From the coefficients above we can deduce that aircrafts with many engines there was less injuries and fatalities or totally uninjured as compared to the opposite.
